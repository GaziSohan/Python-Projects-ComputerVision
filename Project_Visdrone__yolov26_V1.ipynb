```python
# Using pretrained yolo26n.pt model

import cv2
import math
import cvzone

from ultralytics import YOLO
from google.colab.patches import cv2_imshow

model = YOLO('/content/yolo26n.pt')
classnames = model.names
conf_threshold = 0.60
img = cv2.imread('/content/images/afg_car_truck.png')
results = model(img)

for result in results:
    boxes = result.boxes

    for box in boxes:

        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        w = x2 - x1
        h = y2 - y1

        conf = math.ceil(box.conf[0] * 100) / 100

        cls = int(box.cls[0])
        label = classnames[cls]

        if conf >= conf_threshold:

            cvzone.cornerRect(
                img,
                (x1, y1, w, h),
                l=9
            )

            cvzone.putTextRect(
                img,
                text=f'{label} {conf}',
                pos=(max(0, x1), max(35, y1)),
                scale=1.5,
                thickness=2,
                offset=3
            )

cv2_imshow(img)
```
